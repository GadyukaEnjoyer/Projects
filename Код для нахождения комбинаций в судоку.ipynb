```python
def find_combinations(target_sum, start, end, path, result):
    # Базовый случай: если сумма достигла цели и длина комбинации равна 4
    if target_sum == 0 and len(path) == 4:
        result.append(path)
        return
    
    # Если сумма уже превысила целевое значение или длина комбинации больше 4, выходим из функции
    if target_sum < 0 or len(path) > 4:
        return
    
    # Перебираем числа в диапазоне (start, end)
    for i in range(start, end + 1):
        # Рекурсивно ищем комбинации с новым значением
        find_combinations(target_sum - i, i + 1, end, path + [i], result)

def find_all_combinations(target_sum):
    result = []
    find_combinations(target_sum, 1, 9, [], result)
    return result

# Пример использования
target_sum = 22
combinations = find_all_combinations(target_sum)
print("Комбинации чисел, дающих сумму", target_sum, ":", combinations)
```

```python
def find_all_sums():
    sums = {}
    
    # Перебираем все возможные пары чисел от 1 до 9
    for i in range(3, 10):
        for j in range(i+1, 10):  # Начинаем с j=i, чтобы избежать дублирования комбинаций
            current_sum = i + j
            # Если такой суммы еще нет в словаре, создаем новую запись
            if current_sum not in sums:
                sums[current_sum] = [(i, j)]
            else:
                sums[current_sum].append((i, j))  # Добавляем новую пару к существующей сумме
    
    return sums

# Пример использования
sums = find_all_sums()
print("Возможные варианты создания сумм от 1 до 9:")
for total, pairs in sums.items():
    print(f"Сумма {total}: {pairs}")
```

```python
def find_unique_sum_combinations():
    combinations = []
    
    # Перебираем все возможные пары чисел от 1 до 9
    for i in range(1, 10):
        for j in range(i + 1, 10):  # Исключаем повторения
            current_sum = i + j
            # Проверяем, нет ли уже такой суммы в списке комбинаций
            if current_sum not in [sum(pair) for pair in combinations]:
                combinations.append((i, j))  # Добавляем новую комбинацию в список
    
    return combinations

# Пример использования
unique_sum_combinations = find_unique_sum_combinations()
print("Комбинации пар чисел от 1 до 9 с уникальными суммами:")
for pair in unique_sum_combinations:
    print(pair)
```

```python
from itertools import combinations

def divide_numbers():
    numbers = list(range(1, 10))
    valid_combinations = []

    # Получаем все возможные комбинации чисел по 3 в каждой группе
    for group1 in combinations(numbers, 3):
        remaining_numbers_1 = set(numbers) - set(group1)
        for group2 in combinations(remaining_numbers_1, 3):
            remaining_numbers_2 = set(remaining_numbers_1) - set(group2)
            for group3 in combinations(remaining_numbers_2, 3):
                # Проверяем, чтобы сумма каждой группы была кратна 3
                if sum(group1) % 3 == 0 and sum(group2) % 3 == 0 and sum(group3) % 3 == 0:
                    valid_combinations.append((group1, group2, group3))

    return valid_combinations

# Пример использования
valid_divisions = divide_numbers()
print("Варианты разделения чисел от 1 до 9 на три группы по три числа, сумма которых кратна 3:")
for division in valid_divisions:
    print(division)
```
